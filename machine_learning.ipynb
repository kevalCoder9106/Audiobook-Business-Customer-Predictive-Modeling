```python
import numpy as np
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

```python
npz = np.load('train_data.npz')
train_inputs, train_targets = npz['inputs'].astype(float), npz['targets'].astype(int)

npz = np.load('val_data.npz')
val_inputs, val_targets = npz['inputs'].astype(float), npz['targets'].astype(int)

npz = np.load('test_data.npz')
test_inputs, test_targets = npz['inputs'].astype(float), npz['targets'].astype(int)
```

```python
input_size = 10
output_size = 2
hidden_layer_size = 50

model = tf.keras.Sequential([
    Dense(32, activation='relu'),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(output_size, activation='softmax')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
max_epochs = 100
batch_size = 100

model.fit(train_inputs,
         train_targets,
         epochs=max_epochs,
          validation_data=(val_inputs,val_targets),
          verbose=2)
```

```python
# Testing

loss, accuracy = model.evaluate(test_inputs, test_targets)
```
